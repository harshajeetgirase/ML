```python
import matplotlib.pyplot as plt
import numpy as np

X = np.arange(0,10)
y=(X-5)**2


print(X,y)
```

```python
# plot the curve 
plt.plot(X,y)
plt.xlabel("X")
plt.ylabel("F(X)")
plt.show()
```

```python
# gradient descent 

x=0
lr=0.1

plt.plot(X,y)

error=[]

for i in range(50):
    gradient= 2*(x-5)
    x=x-gradient*lr
#     print(x)
    y=(x-5)**2
    error.append(y)
    plt.scatter(x,y)
```

```python
# plotting the error 
plt.plot(error)
```
